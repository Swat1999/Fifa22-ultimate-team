# FIFA 22 Ultimate Team Analysis and Maker

## In this Analysis, we will be analyzing the dataset for the new FIFA 22 Ultimate team video game and create our very own Playing 11 based on the stats on a dataframe both worst and the best team

## We need a formation first, cause in football (soccer) we need formation in the field to play the game hence we will take the formation of "1-3-4-3" that means
### 1 = Goalkeeper.
### 3 = 3 Defenders.
### 4 = 4 Mid Fielders.
### 3 = 3 Attackers or forward.

### This formation is pretty balanced and based on this formation, there will be different sub positions so hence we will be analyzing those and get the best team out of it

### Another interesting thing we will do is set price so hence we can keep ourself within a budget to pick players cause in  reality we cannot afford everyone and that is the same in the game as well

**We will use the picture above to set our formation so from the dataset we need
** G, LB, 2 CBs, LM, DM, AM, RM, LF, CF, RF

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fut22.csv')
```

```python
df
```

```python
df_drop = df.drop(['player_image', 'club_image', 'league_image', 'nationality_image'], axis=1)
```

```python
df_drop
```

```python
df_drop.isnull().sum()
```

```python
df_drop.dtypes
```

```python
cols_sum = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physicality']
```

```python
df_drop['total_points'] = df[cols_sum].sum(axis=1)
```

```python
df_drop['average_points'] = df_drop['total_points'] / 6
```

```python
df_drop
```

```python
df_drop.dtypes
```

```python
df_drop['average_points'] = df_drop['average_points'].astype(int)
```

```python
df_drop
```

```python
df_new = df_drop
```

```python
df_new['average_points'].value_counts()
```

```python
conditions = [
    (df_new['average_points'] > 50) & (df_new['average_points'] < 60),
    (df_new['average_points'] > 60) & (df_new['average_points'] < 70),
    (df_new['average_points'] > 70) & (df_new['average_points'] < 80),
    (df_new['average_points'] >= 80)
]

prices = ['350', '500', '750', '1000']
```

```python
df_new['Prices'] = np.select(conditions, prices)
```
